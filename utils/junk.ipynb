```python
from pprint import pprint
from comprehend_detect import ComprehendDetect
import pandas as pd
import boto3
from botocore.exceptions import ClientError

df = pd.read_csv("data/pii_test_faker.csv", nrows=50)
comp_detect = ComprehendDetect(boto3.client("comprehend","us-east-1"))

def detect_names(df, rows=5):
    df_names = pd.DataFrame()
    for i in range(rows):
        for col in df.columns:
            name_list = comp_detect.detect_entities(str(df[col][i]), 'en')
            for name in name_list:
                if name['Type'] == "PERSON":
                    name['field'] = col
                    name['row'] = i+1
                    print(name)
                    df_names = pd.concat([df_names,pd.DataFrame([name])])
    pprint(df_names.head())

    return df_names
detect_names(df,50)
```
